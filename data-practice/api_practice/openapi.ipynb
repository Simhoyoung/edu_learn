```python
item = response_python["response"]["body"]["items"]["item"]
pprint(item)
```

```python
# 온도를 나타내는 데이터 모두 출력
for i in item:
    category = i["category"]
    if category == "T1H":
        print(i)
```

```python
# 최고 온도, 해당 날씨 데이터 출력
max_temp = -99

# 최고기온일 때의 날시 데이터
max_weather = {}

for i in item:
    category = i['category']
    #기온 데이터
    if category == 'T1H':
        # 기온이 몇
        # 기온 데이터 추출, 정수(int) 변환 후 저장
        fcstValue = int(i['fcstValue'])
        # 최고 기온 저장값보다 현재 기온이 크면
        if max_temp < fcstValue:
            # 최고 기온 저장값을 현재 기온으로 갱신
            max_temp = fcstValue
            # 최고 기온 날씨 저장값을 현재 날씨 데이터로 갱신
            max_weather = i

print(max_temp)
print(max_weather)
```
