```python
# tips 데이터를 활용한 산점도
# 모듈 불러오기
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터셋 불러오기
tips_df = sns.load_dataset("tips")
```

```python
tips_df.info()
```

```python
tips_df.head()
```

```python
# 산점도
# 2개의 연속형 데이터를 기반으로

# total_bill에 따른 tip을 산점도 시각화
sns.scatterplot(
    data=tips_df,
    x="total_bill",  # x축에 사용할 열 이름
    y="tip",  # y축에 사용할 열 이름
)

# 그래프를 출력
plt.show()
```

```python
# hue 매개변수
# 점(포인트)를 구분하는 범주형 데이터

sns.scatterplot(
    data=tips_df,
    x="total_bill",
    y="tip",
    # hue="size",# 포인트를 구분하는 범주형 데이터
    hue="day"# 포인트를 구분하는 범주형 데이터
) 

plt.show()
```
