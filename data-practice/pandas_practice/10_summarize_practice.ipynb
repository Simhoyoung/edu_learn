```python
import pandas as pd

df = pd.read_csv("../data/vgsales.csv")
```

```python
# 장르별 세계 판매량 합계를 출력한다.

grouped = df.groupby("genre")["total_sales"].sum()
grouped.sort_values(ascending=False)
```

```python
# 플랫폼별 세계 판매량 평균을 출력한다.

# 가장 평균이 높은 플랫폼의 값
df.groupby("console")["total_sales"].mean().sort_values(ascending=False).iloc[0]
# 가장 평균이 높은 플랫폼
df.groupby("console")["total_sales"].mean().sort_values(ascending=False).index[0]
```

```python
# 개발사별 게임 출시 개수를 출력
grouped = df.groupby("developer")["title"].count()

# 가장 많은 개발사(행 이름)
grouped.sort_values(ascending=False).index[1]
```

```python
# 개발사별 게임 출시 개수를 출력한다.

df.groupby("publisher")["title"].count()
```

```python
# 플랫폼별 세계 판매량, 북미 판매량, 일본 판매량, 유럽 및 지역 아프리카 판매량,
#  기타 지역 판매량의 합계와 평균을 출력한다.

func = ["sum", "mean"]
grouped = df.groupby("console")
grouped.agg(
    {
        "total_sales": func,
        "na_sales": func,
        "jp_sales": func,
        "pal_sales": func,
        "other_sales": func,
    }
)

# 동일한 계층 컬럼이 중복 들어갈때, 한번만 사용
grouped[[
    "total_sales",
    "na_sales",
    "jp_sales",
    "pal_sales",
    "other_sales",
]].agg(func)
```

```python
# 평론가 점수의 평균이 9.5가 넘는 게임 개발사를 선택해서 출력한다.
# df["critic_score"] = df["critic_score"].fillna()
group_rows = df.groupby("developer")


def greater_score(group):
    group_rows = group["critic_score"].mean()
    # df["critic_score"].fillna(group_row["critic_score"].mean())
    return group_rows >= 9.5


result = group_rows.filter(greater_score, dropna=True)[["developer", "critic_score"]]
# result.fillna()
result.drop_duplicates()

# 결측치 계산 X
# [1,1,1,1,NaN] : (1+1+1+1) / 4
```
