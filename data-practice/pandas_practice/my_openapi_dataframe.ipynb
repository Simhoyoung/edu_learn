```python
# 데이터 타입 변환
df["so2Value"] = pd.to_numeric(df["so2Value"], errors='coerce')
df["coValue"] = pd.to_numeric(df["coValue"], errors='coerce')
df["pm10Value"] = pd.to_numeric(df["pm10Value"], errors='coerce')
df["so2Grade"] = pd.to_numeric(df["so2Grade"], errors='coerce')

# NaN 값 처리 (필요 시)
df.dropna(subset=["so2Value", "coValue", "pm10Value", "so2Grade"], inplace=True)

# 조건 필터링
condition1 = df["so2Value"] < 0.004
condition2 = df["coValue"] == 0.5
condition3 = df["pm10Value"] == 35
condition4 = df["so2Grade"].isin([2.0])

df.loc[(condition1 & condition2) | condition3 | ~condition4]

```
