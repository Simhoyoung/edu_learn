```python
import pandas as pd
```

```python
df2 = pd.read_csv("../data/titanic_2.csv")
df3 = pd.read_csv("../data/titanic_3.csv")
```

```python
df2.head()
df3.head()
```

```python
df = pd.concat([df2, df3], ignore_index=True)
df
```

```python
unique_df = df.drop_duplicates()
unique_df
# 중복 데이터 삭제 후 너무 많은 데이터가 날라가면
# 분석이 어려울 수 있음, 새로 만들거나 새로 수집해야 할 수 도 있 다 구 웅웅
```

```python
unique_df.head()
```

```python
unique_df = unique_df.drop(columns=["alive", "embarked", "class", "deck", "who"])
```

```python
# 요금이 0인 행을 선택
condition = unique_df["fare"] == 0

# 0인 fare 값을 10으로 수정
unique_df.loc[condition, "fare"] = 10
```

```python
unique_df.sort_values(by="fare")
```

```python
# 나이가 1보다 작은 행을 선택
condition = unique_df["age"] < 1
unique_df.loc[condition, "age"] = 1
unique_df.sort_values(by="age")
```
