```python
# 열 이름을 수정할 수 있다
# 단, 리스트 길이와 기존 열 개수가 동일해야 한다.

df.columns = [
    'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    'pH', '산도', 'alcohol', 'quality'
    ]
```

```python
# 산도를 기준으로 오름차순 정렬
# by : 어떤 열을 기준으로 정렬?
# asending : 오름차순
df.sort_values(by='pH', ascending=True)
```

```python
# qulity를 기준으로 내림차순으로 정렬
df.sort_values(by='quality', ascending=False)
```

```python
# 주의사항
# 정렬한 결과가 원본 데이터 프레임에 적용 X
# 기존 데이터프레임은 변화가 없음
# 새로운 데이터프레임을 생성하는 경우, 새로운 변수에 저장을 해서 사용해야한다용용
sorted_df = df.sort_values(by="quality", ascending=True)
sorted_df
```

```python
# quality, alcohol 기준 오름차순 정렬
df.sort_values(by=["quality", "alcohol"], ascending=[True,True])

```

```python
# quality, alcohol, residual sugar 기준 오름차순 정렬
df.sort_values(by=["quality", "alcohol", "residual sugar"])
```

```python
# quality 기준 내림, alcohol 기준 오름, residual sugar 오름
df.sort_values(by=["quality", "alcohol", "residual sugar"], ascending=[False,True,True])
```

```python
# sort_values(by=[]) : 열의 값을 기준으로 정렬
# sort_index() : 행의 이름(인덱스)을 기준으로 정렬

# 행의 이름은 인덱스(index)
# 열의 이름은 열의 이름
```
