```python
import pandas as pd

df = pd.read_csv("../data/car_evaluation.csv")
```

```python
# 데이터프레임의 구조를 파악하기 위한 코드를 3개 이상 작성한다.
df.head()
df.info()
```

```python
# 각 열에 존재하는 고유값의 개수를 출력한다.
df.nunique()
```

```python
# 아래 열에 존재하는 고유값 목록을 출력한다.
# number of doors
# number of persons
# price
# class
num_door_df = df["number of doors"].unique()
num_person_df = df["number of persons"].unique()
price_df = df["price"].unique()
class_df = df["class"].unique()
print(num_door_df, num_person_df, price_df, class_df)
```

```python
# 아래 열에 존재하는 고유값들의 비중을 출력한다.
# number of doors
# number of persons
# price
# class
df["number of doors"].value_counts(normalize=True)
df["number of persons"].value_counts(normalize=True)
df["price"].value_counts(normalize=True)
df["class"].value_counts(normalize=True)
```

```python
# 아래 열들을 범주형 데이터 타입으로 변환한다. 또한, 아래 대소관계를 지켜야한다.
# number of doors : 5more > 4 > 3 > 2
# number of persons : more > 4 > 2
# price: vhigh > high > med > low
# 1. 범주형 데이터 목록
# price의 순서
orders = ["low", "med", "high", "vhigh"]
# 2. 변환할 열
column = df["price"]

# 범주형 자료형데이터 생성
df["price"] = pd.Categorical(
    values=column,
    categories=orders,
    ordered=True,
)
```

```python

# 1. 범주형 데이터 목록
# class의 순서
orders = ["unacc", "acc", "good", "vgood"]
# 2. 변환할 열
column = df["class"]

# 범주형 자료형데이터 생성
df["class"] = pd.Categorical(
    values=column,
    categories=orders,
    ordered=True,
)
```

```python
# class 기준 내림차순 price 기준 오름차순으로 정렬하고, 출력한다.

# sort_values() 활용 정렬
# 조건 열이 2개 이상은 by에 리스트를 입력
df.sort_values(by=["class", "price"], ascending=[False, True])
```

```python
# price 기준 오름차순 class 기준 내림차순으로 정렬하고, 출력한다.

# sort_values() 활용 정렬
# 조건 열이 2개 이상은 by에 리스트를 입력
df.sort_values(by=["class", "price"], ascending=[True, False])
```
