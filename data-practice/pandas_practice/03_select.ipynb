```python
df = pd.read_csv("../data/Car.csv")
```

```python
# 열 선택
# df['name']
# df['fuel']

# 2개 이상의 열을 선택할 때는
# 열들을 리스트에 넣어서 선택해야 한다
df[["name", "fuel"]]
```

```python
# 행 슬라이싱
# 시작부터 끝까지
df[0:]
# 위치를 기반으로 슬라이싱
df[5:10]
# 사용 많이 안한다고함
```

```python
df.head()

# 모든 행을 선택하는데 첫번째 두번째 세번째 열을 선택해서 출력
df.iloc[:, 0:3]

# 5:10 행을 선택하는데 0:6 열을 선택해서 출력
# df.iloc[5:10,0:6]
```

```python
# .loc[] : 레이블 기반 선택
# .loc[행, 열]

# 모든 행을 선택
df.loc[:]

# 행 이름이 0
df.loc[0]

# 행 이름이 0, 1, 3
df.loc[[0, 1, 3]]

# 슬라이싱, **종료 레이블이 포함
df.loc[0:10]

# 단일 열(name) 선택
df.loc[:, "name"]

# 복수 열(name, seller_type)
df.loc[:, ["name", "seller_type"]]

# 행과 열 선택 조합
df.loc[[0, 3, 5], ["name", "selling_price"]]
df.loc[0:10, ["name", "selling_price"]]
```

```python
# 1. selling_price 열 기준으로 내림차순 정렬
# 2. 결과를 새로운 변수(sorted_df)에 저.장.
sorted_df = df.sort_values(by="selling_price", ascending=False)
```

```python
# 3. sorted_df 에 대해서 인덱스 기반 선택
# .iloc[0]
sorted_df.iloc[0]
# 현재 데이터프레임의 '현재 상태' 순서 기준
```

```python
# 4. sorted_df 에 대해서 레이블 기반 선택
# .loc[0]
sorted_df.loc[0]
# 전체 기준이라 정렬 데이터가 의미가 없어짐짐
```

```python
# 조건식
# 열을 선택해서 해당 열의 개별 데이터와 값을 비교하는 식
# 판매 가격(selling_price) 100000 보다 작은

# 1. 열 선택
# 2. 조건 작성
# 3. 비교 연산자 작성
condition = df["selling_price"] < 100000

# 4. loc를 사용해서 데이터 선택(필터링)
df.loc[condition]
```

```python
# 주행거리(km_driven)이 100000 보다 큰 데이터를 필터링
km_driven = df["km_driven"] > 100000

# name열과 selling_price 열 선택
df.loc[km_driven, ["name", "selling_price"]]
```

```python
# 연료(fuel) "Diesel"인 데이터를 필터링
diesel_fuel = df["fuel"] == "Diesel"
df.loc[diesel_fuel]
```

```python
# 판매유형(seller_type) "Dealer"가 아닌 데이터 필터링
dealer_seller = df["seller_type"] != "Dealer"
df.loc[dealer_seller]
```

```python
# 주행거리가 100000 보다 작은
condition1 = df["km_driven"] < 100000

# 판매가격이 100000 보다 작은
condition2 = df["selling_price"] < 100000

# 두 조건을 모두 만족하는 데이터를 선택
df.loc[condition1 & condition2]
```

```python
# 주행거리가 100000 보다 작은
condition1 = df["km_driven"] < 100000

# 판매가격이 100000 보다 작은
condition2 = df["selling_price"] < 100000

# 운전방식(transmission)이 Automatic
condition3 = df["transmission"] == "Automatic"
df.loc[condition1 & condition2 & condition3]
```

```python
# fuel 가 Diesel 이거나 transmission 이 Automatic
condition1 = df["fuel"] == "Electric"
condition2 = df["transmission"] == "Automatic"
df.loc[condition1 | condition2]
```

```python
# year 가 2018 보다 크면서
# fuel 가 Petrol 또는 Diesel 인 데이터 선택
condition1 = df["year"] > 2018
condition2 = df["fuel"] == "Petrol"
condition3 = df["fuel"] == "Diesel"

df.loc[condition1 & (condition2 | condition3)]
```

```python
# isin
# 연료가 Petrol, Diesel, LPG
condition = df["fuel"].isin(["Petrol", "Diesel", "LPG"])

# True
df.loc[condition]

# False
df.loc[~condition]
```

```python
# year가 2020, 2018, 2016, 2014 인 데이터 선택
condition = df["year"].isin([2020, 2018, 2016, 2014])

df.loc[condition]
```
