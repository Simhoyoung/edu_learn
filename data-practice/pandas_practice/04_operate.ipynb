```python
# 04_operate.ipynb
import pandas as pd

# 데이터 프레임 생성
df = pd.DataFrame(
    {
        "A": [1, 2, 3],
        "B": [4, 5, 6]
    }
    )
```

```python
# 행 이름이0의 A 열 데이터 수정
df.loc[0,"A"] = 100
df
```

```python
# 행 이름이 0,1의 B 열 데이터 수정
df.loc[[0,1],"B"] = [10,20]
df
```

```python
# A열의 데이터를 모두 10으로 수정
df.loc[0,"A"] = 110
df
```

```python
# B열의 데이터 중 10보다 작은 데이터를 10으로 수정
# 1. 10보다 작은 B열의 데이터를 필터링
condition = df["B"] < 10

# 필터링 후 값 수정정
df.loc[condition, "B"] = 10
df
```

```python
# 데이터 수정하는 경우

# 1. 이상치를 수정
#   말도안되는 값, 우선 찾고 제거하거나 수정

# 2. 값의 통일성
#   지역 열이 있는데 어떤 데이터는 서울, 어떤 데이터는 서울특별시~
```

```python
# 열 'A' 이름을 '가'로 수정
# 기존 데이터프레임을 수정하지 않는다다
df = df.rename(columns={"A": "가"})
df
```

```python
# 행 이름 0을 1로 수정
df = df.rename(index={0:1})

# 행 이름 1을 선택?
df.loc[1]
```

```python
# 데이터 타입을 수정
df = pd.DataFrame({"A": [1, 2, 3], "B": [4.5, 5.5, 6,4]})

# astype()
# 기존 데이터프레임을 수정하지 않는다
df["A"] = df["A"].astype(float)
df.info()
```

```python
# 행 추가하기
# concat() : 2개 이상의 데이터프레임을 합치는 함수

df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
new_df = pd.DataFrame({"A": [7], "B": [8]})

# ignore_index : 합친 후 행 이름을 초기화한다
df = pd.concat([df, new_df], ignore_index=True)
df
```

```python
# 존재하지 않는 열에 대해 값을 저장하면?
# 새로운 열을 생성하고, 값을 저장장
df["C"] = 10
df
```

```python
# B열의 데이터가 짝수라면
condition = df["B"] % 2 == 0

df["D"] = condition
df
```

```python
import numpy as np

condition = df["B"] % 2 == 0
df["E"] = np.where(condition, "짝수", "홀수")
df
```

```python
# drop()
# drop(columns=[]) 열 삭제
# drop(index=[]) 행 삭제

# 열 삭제, 기존의 데이터프레임은 수정 X
df.drop(columns=["이름", "직업"])

# 행 삭제
# 점수가 60 미만인 행을 삭제
condition = df['점수'] < 60

# 행을 선택(필터링), index만 저장
drop_index = df.loc[condition].index

df.drop(index=drop_index)
```

```python
df.duplicated()
# 처음 제외하고 찾아줌 3,16
df.loc[[1, 3, 16]] 
```

```python
# 이름, 직업이 같은 경우 중복 처리
# 일부 열에 대해서만 중복 처리
df.loc[df.duplicated(subset=["이름", "직업"])]
```

```python
# 중복 데이터 삭제
# 모든 열이 동일한 중복 데이터 삭제
# 기존 데이터프레임 수정 X
df.drop_duplicates()

# keep 인자
df.drop_duplicates(keep="last")

# subset 인자
# 이름, 직업, 지역이 중복인 경우 삭제
df.drop_duplicates(subset=["이름", "직업", "지역"])
```
