```python
#

# 부서 당 인원 수 출력
# 1. 그룹화
grouped = df.groupby("부서")

# 2. "이름" 열에 집계 함수 count()
grouped["이름"].count()

# 부서의 인원 수가 2명보다 많은 부서를 필터링
# 1. 그룹화
grouped = df.groupby("부서")


# 조건 함수
# 매개 변수 : 그룹화한 데이터의 개별 행
def greater_count_two(group):
    count = group["이름"].count()
    # count가 2보다 크면 결과값이 True
    if count > 2:
        return True
    # count가 2보다 작으면 결과값이 False
    if count < 2:
        return False


# 2. 필터링
grouped.filter(greater_count_two)
```

```python
# 평균 연봉이 6000 이상인 부서만 필터링
# 1. 그룹화
grouped = df.groupby("부서")


# 2. 조건 함수
# 매개변수 : 부서를 기준으로 그룹화된 행 데이터터
def greater_salary(group):
    # 평균 연봉
    mean_salary = group["연봉"].mean()

    # 평균 연봉이 6000 이상이면
    if mean_salary > 6000:
        return True
    # 아니라면
    else:
        return False


# 3. filter() 적용
grouped.filter(greater_salary)
```

```python
# 평균 연봉이 6000 이상인 부서만 필터링
# 1. 그룹화
# group_rows : 그룹화된 데이터 행의 모음
group_rows = df.groupby("부서")

# 2. 조건함수
def greater_salary(group_row):
    mean_salary = group_row["연봉"].mean()

    return mean_salary >= 6000

# 3. 필터링
group_rows.filter(greater_salary)

```

```python
# 연봉 열에 평균, 중위값
# df["연봉"].mean()
# df["연봉"].median()
# 연봉의 중위값과 평균
df["연봉"].agg(["median", "mean"])

# 연봉의 최소값과 최대값
df["연봉"].agg(["min", "max"])

# 이름의 개수
# 연봉의 평균, 중위값
# 나이의 최소, 최대, 평균
df.agg(
    {
        "이름": "count",
        "연봉": ["median", "mean"],
        "나이": ["min", "max", "mean"],
    }
)
```

```python
# 그룹화된 데이터에도 다중 집계 함수 agg()
grouped = df.groupby("지역")

result = grouped.agg(
    {
        "이름": "count",
        "연봉": ["median", "mean"],
        "나이": ["min", "max", "mean"],
    }
)

# 다중 집계함수의 결과 :
# 지역이 index 역할
# 컬럼이 계층 구조를 가진다.

# 연봉 - mean 선택하고 싶으면?
result[("연봉", "mean")]

# 지역 > 나이 - min,max 선택
result.loc[:,("나이",["min","max"])]
```
