```python
import seaborn as sns
import matplotlib.pyplot as plt

# iris : 붓꽃
iris_df = sns.load_dataset("iris")

iris_df.head()
iris_df.info()
# species : 품종 -범주형
# sepal : 꽃받침 길이/너비
# petal : 꽃잎 길이/너비
```

```python
sns.barplot(
    data=iris_df,
    x="species",
    y="sepal_length",
    estimator="max" #적용할 집계 함수
)

plt.show()
```

```python
# 펭귄 시각화 실습

penguin_df = sns.load_dataset("penguins")
penguin_df.info()
```

```python

sns.barplot(
    x="island",
    y="body_mass_g",
    data=penguin_df,
    hue="sex"
)

plt.show()
```

```python

sns.barplot(
    x="island",
    data=penguin_df,
    hue="species"
)

plt.show()
```

```python
sns.boxplot(
    data=penguin_df,
    y="body_mass_g",
    hue="species"
)

plt.show()
```
