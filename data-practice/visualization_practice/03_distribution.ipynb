```python
import seaborn as sns
import matplotlib.pyplot as plt

penguin_df = sns.load_dataset("penguins")
penguin_df.head()
penguin_df.shape
```

```python
sns.histplot(
    data=penguin_df,
    x="bill_length_mm",
    bins=15,  # 히스토그램 bin(막대의 개수) 조절
    hue="species",
    multiple="fill", 
    # multiple : layer 기본값 , 겹쳐서 표현
    # stack : 쌓아서 구분해서 표현
    # dodge : 옆으로 나란히 구분해서 표현
    # fill : 각 범주의 막대 높이를 비율로 표현현
)

plt.show()
```

```python
# 펭귄 종에 따른 날개 길이 분포 KDE Plot
sns.kdeplot(
    data=penguin_df,
    x="flipper_length_mm",
    hue="species", # 종에 따른 구분
    fill=True,
    bw_adjust=.5, # 대역폭, 커지면 곡선 부드러워짐, 작으면 날카로워짐짐
)

plt.show()
```

```python
# 분포형 그래프들의 부모 함수
sns.displot(
    data=penguin_df,
    x="body_mass_g",
    hue="species",
    kind="kde", # 그래프종류,
    rug=True,
)

plt.show()
```
