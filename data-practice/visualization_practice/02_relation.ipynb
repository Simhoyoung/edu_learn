```python
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 실습 데이터 tips
tips_df = sns.load_dataset("tips")

tips_df.head()
```

```python
tips_df["time"].unique()
```

```python
# 시각화 요소
# 데이터, x축, y축, 구분데이터(hue)

sns.scatterplot(
    data=tips_df,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
# 항공기 탑승객 수
flight_df = sns.load_dataset("flights")
flight_df.head()
flight_df.info()
```

```python
# line plot(data, x, y)
sns.lineplot(data=flight_df,x="year",y="passengers")
plt.show()

# 중앙선 연도의 평균
# (하늘색)신뢰구간 - 표본데이터(일부 추출데이터터)/모수(전국 데이터 - 사실상 어렵다)
```

```python
# 월별 승객수
sns.lineplot(data=flight_df,x="month", y="passengers")

plt.show()
```

```python
# FacetGrid
# Grid : 격자 형태로 여러개 그래프(서브 플롯) 시각화해서
# 관계를 확인한다

tips_df = sns.load_dataset("tips")
```

```python
grid = sns.FacetGrid(
    data=tips_df,
    col="day", # 구분을 위한 범주형데티어(day, smoker, time,,,,)
)

# 그리드에 그래프를
grid.map(plt.hist, "total_bill")
plt.show()
```
