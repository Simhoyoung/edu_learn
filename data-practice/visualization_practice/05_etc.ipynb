```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = sns.load_dataset("titanic")
titanic_df.info()
```

```python
#    
#   
numeric_df =titanic_df.select_dtypes(include=["float64", "int64"])

#    
matrix =numeric_df.corr()

# heatmap
sns.heatmap(
    matrix,
    vmin=.1,
    vmax=1.0,
    annot=True,
    cmap="plasma"
)
plt.show()
```
